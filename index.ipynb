```python
import numpy as np 
import pandas as pd
import sklearn.datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the data from sklearn
breast_cancer_dataset =  sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#loading the data to a pandas dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)

data_frame.head()
```

```python
data_frame['label'] = breast_cancer_dataset.target
```

```python
#printing the last 5 rows of the dataframe
data_frame.tail()
```

```python
data_frame.shape
```

```python
data_frame['label'].value_counts()

# value 1 = Benign
# value 0 = Malignant
```

```python
data_frame.groupby('label').mean()
```

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
# splitting the data into training data and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Training data accuracy = ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on test data = ', test_data_accuracy)
```

```python
# instead of logistic regression, lets say we wanted to use neural networks, then..
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
# import tensorflow and keras

import tensorflow as tf 
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# create the neural network layers
model2 = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
model2.compile(optimizer='adam', loss = 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model2.fit(X_train_std, Y_train, validation_split = 0.1, epochs=10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python
# accuracy of the model on the test data
loss, accuracy = model2.evaluate(X_test_std, Y_test)
print(accuracy)
```
